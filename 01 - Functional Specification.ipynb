# Objectives

Using the history of pesticides and microbeads, Sellers et.al. (2017) illustrate the ability of awakened public outrage to draw attention to long-dormant concerns about chemicals, triggering regulatory inquiry and changes in policy. In their words: "the quiet cautions of scientists rested in journals on library shelves for decades until a passionate storyteller captured the public's imagination through social media. The resulting outcry led to retail regulation and legislative action.". 

Sellers et.al. advocate for vigilant, pro-active management of product exposures to public sentiment through monitoring "leading" and "lagging" indicators, assessing impact of potential shifts in sentiment or regulatory action and responding appropriately to potential risks. Leading indicators include "social media, non-governmental organization (NGO) activity, customer inquiries, scientific research, legislative trends", while lagging indicators include "scientific research, lawsuits, listing of a chemical on a 'watch list' by a retailer or NGO, shareholder resolutions, proposed legislation". 

The scale of text information generated by the sources of interest can be daunting to review and analyze. An efficient method of screening and prioritizing publications, statements and other text-based information is required to efficiently guide product stewards toward the most urgent issues that may impact business. This exercise is intended to develop a statistical method as a solution to this problem.

# Scope

This exercise will explore public sentiment and regulatory action relating to: 
1) Microbeads
2) PFOS/PFOA
These two families of chemicals have been identified in recent years as emerging contaminants and would provide valuable, timely case studies for identifying trends in sentiment and policy. 

This exercise will explore the potential for analyzing information from: 
1) social media
2) websites
3) search engines

The text obtained from these sources will be analyzed using statistics based on relevance, sentiment and frequency to evaluate public sentiment and regulatory action. A framework for evaluating public sentiment and regulatory action will be developed during the course of the study.
